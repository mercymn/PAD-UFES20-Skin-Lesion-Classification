```python
import pandas as pd
from sklearn.preprocessing import KBinsDiscretizer, OrdinalEncoder
```

```python
df = pd.read_pickle("../data/skin-lesion-data/binary.pickle")
```

```python
df.describe(include="all")
```

# Drop irrelevant features

```python
df = df.drop("patient_id", axis=1)
df = df.drop("lesion_id", axis=1)
df = df.drop("img_id", axis=1)
df
```

```python
df.select_dtypes(include=["int", "float"]).columns
```

# Binning



## Age 3 bins

1. Young
2. Adult
3. Elderly

```python
labels = ["young", "adult", "elderly"]
discretizer = KBinsDiscretizer(n_bins=len(labels), encode="ordinal", strategy="uniform")
df["age"] = discretizer.fit_transform(df[["age"]]).astype(int)
df["age"]
```

## Fitspatrick 6 bins

1. I
2. II
3. III
4. IV
5. V
6. VI

```python
df["fitspatrick"].unique()
```

```python
df["fitspatrick"] = df["fitspatrick"].astype(int)
df["fitspatrick"]
```

## Diameter 3 bins

1. Short
2. Medium
3. Long

```python
labels = ["short", "medium", "long"]
```

```python
discretizer = KBinsDiscretizer(n_bins=len(labels), encode="ordinal", strategy="uniform")
df["diameter_1"] = discretizer.fit_transform(df[["diameter_1"]]).astype(int)
df["diameter_1"]
```

```python
discretizer = KBinsDiscretizer(n_bins=len(labels), encode="ordinal", strategy="uniform")
df["diameter_2"] = discretizer.fit_transform(df[["diameter_2"]]).astype(int)
df["diameter_2"]
```
