## Dataset is really unbalanced

```python
df["diagnostic"].value_counts()
```

## Dataset is more balanced like this

```python
malignant = ["BCC", "MEL", "SCC"]
diagnostic_amount = df["diagnostic"].value_counts()
cancerous_amount = {}
for d, v in diagnostic_amount.items():
    if d in malignant:
        cancerous_amount["malignant"] = cancerous_amount.get("malignant", 0) + v
    if not d in malignant:
        cancerous_amount["benign"] = cancerous_amount.get("benign", 0) + v
cancerous_amount
```

```python
df["malignant"] = df["diagnostic"].isin(malignant)
df
```

```python
df = df.drop("diagnostic", axis="columns")
```
