```python
import os
import random
import numpy as np
import imageio.v3 as iio
from pathlib import Path
import torch
import torchvision
import torch.nn as nn
import torch.optim as optim
```

```python
IMAGES_DIR = Path("../data/skin-lesion-data/images/train/")
```

```python
images_path = os.listdir(IMAGES_DIR)
random.shuffle(images_path)
len(images_path)
```

```python
bath_size = 40
image_size = (200, 200)
```

```python
batches = [images_path[i: i + bath_size] for i in range(0, len(images_path), bath_size)]
```

```python
model = nn.Sequential(
    nn.Flatten(),
    torchvision.ops.MLP(
        in_channels=200*200,
        hidden_channels=[128, 128, 1],
        activation_layer=nn.ReLU
    )
)
criterion = nn.BCEWithLogitsLoss()
optimizer = optim.Adam(model.parameters(), lr=0.001)
# example labels for one batch
y = torch.ones((bath_size, 1)).float()
```

```python
def train(epochs: int):
    for epoch in range(epochs):
        optimizer.zero_grad()
        for batch in batches[:1]:
            batch_tensors = []
            for image in batch:
                x = iio.imread(IMAGES_DIR / image)

                # Remove alpha channel if RGBA
                if x.shape[-1] == 4:
                    x = x[..., :3]

                # Grayscale conversion
                gray = np.dot(x, [0.299, 0.587, 0.114])

                gray_t = torch.from_numpy(gray).float()

                batch_tensors.append(gray_t.flatten())
            t = torch.stack(batch_tensors)
            p = model.forward(input=t)
            print(y.shape)
            print(p.shape)
            loss = criterion(p, y)
            loss.backward()
            optimizer.step()

            print(f"Epoch {epoch+1}, Loss: {loss.item():.4f}")
            print("Prediction: ", p[0])




print("Batch tensor shape:", t.flatten(start_dim=-2, end_dim=-1).shape)
```
